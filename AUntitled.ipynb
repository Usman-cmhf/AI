```python
import os
import sagemaker
from sagemaker import get_execution_role

sagemaker_session = sagemaker.Session()

role = "arn:aws:iam::346679108929:role/service-role/AmazonSageMaker-ExecutionRole-20210302T115667"
region = sagemaker_session.boto_session.region_name
```

```python
role
```

```python
training_data_uri = 's3://sagemaker-sample-data-{}/tensorflow/mnist'.format(region)
training_data_uri
```

```python
#!pygmentize 'mnist.py'

# TensorFlow 2.1 script
!pygmentize 'mnist2.0.py'
```

```python
from sagemaker.tensorflow import TensorFlow
mnist_estimator2 = TensorFlow(entry_point='mnist2.0.py',
                             role=role,
                             instance_count=2,
                             instance_type='ml.m5.4xlarge',
                             framework_version='2.1.0',
                             py_version='py3',
                             distribution={'parameter_server': {'enabled': True}})
```

```python
mnist_estimator2.fit(training_data_uri)
```

```python
predictor2 = mnist_estimator2.deploy(initial_instance_count=1, instance_type='ml.t2.medium')
```

```python
import numpy as np

!aws --region {region} s3 cp s3://sagemaker-sample-data-{region}/tensorflow/mnist/train_data.npy train_data.npy
!aws --region {region} s3 cp s3://sagemaker-sample-data-{region}/tensorflow/mnist/train_labels.npy train_labels.npy

train_data = np.load('train_data.npy')
train_labels = np.load('train_labels.npy')
```

```python
predictions2 = predictor2.predict(train_data[:50])
for i in range(0, 50):
    prediction = np.argmax(predictions2['predictions'][i])
    label = train_labels[i]
    print('prediction is {}, label is {}, matched: {}'.format(prediction, label, prediction == label))
```

```python
predictor2.delete_endpoint()
```
